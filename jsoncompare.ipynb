```python
from cmipld.utils.similarity_fingerprint import JSONSimilarityFingerprint
from cmipld.utils.ldparse import strip_fields, name_extract

# Prepare data
clean_data = strip_fields(data)
data_dict = name_extract(clean_data)

# Setup
fp = JSONSimilarityFingerprint(include_keys=False)
fp.load_from_dict(data_dict)
fp.embed()
fp.compute_similarity()

# Compute all (automatically uses self.data_dict)
fp.compute_pairs(threshold=0.85)
fp.compute_clusters(n_clusters=5)
fp.compute_outliers(threshold=0.5)
fp.compute_diffs(threshold=0.85)  # Gets data from data_dict!

# Access results
print(fp.pairs)      # [[file1, file2, sim], ...]
print(fp.clusters)   # {0: [files], 1: [files], ...}
print(fp.outliers)   # [file1, file2, ...]

# Output
markdown = fp.pairs_to_markdown()
diff = fp.get_diff('file1.json', 'file2.json')
results_dict = fp.to_json()  # Just returns dict, no file saving!

# If you want to save JSON:
import json
with open('results.json', 'w') as f:
    json.dump(results_dict, f, indent=2)
```
