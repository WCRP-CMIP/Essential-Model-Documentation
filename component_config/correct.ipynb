```python
import glob,json
js = glob.glob('old/*.json')
import cmipld
```

```python
def correct(f):
    j = json.load(open(f,'r'))
    component = j['model_component'].replace('.','-')
    realm = cmipld.get('emd:model_component/'+component)['component']['@id']
    
    v = j['vertical_computational_grid']
    h = j['horizontal_computational_grid']

    id= f'{realm}_{component}_{h}_{v}'

    print(id)

    jsc = {'validation_key': id,
            'ui_label': id,
            'description': j['description'],
            'horizontal_computational_grid': h,
            'model_component': component,
            'vertical_computational_grid':v,
            '@context': '_context',
            '@type': ['emd', 'wcrp:component_config', 'esgvoc:ComponentConfig'],
            '@id': id}


    json.dump(jsc,open(id+'.json','w'))

    return jsc





correct(js[6])
```

```python

fail = []

for i in js:
    try:
        correct(i)
    except:
        fail.append(i)
```
